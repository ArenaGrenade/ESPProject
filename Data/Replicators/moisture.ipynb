```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.array([100,99,93,90,88,86,84,82,79,75,72,69,67])
from scipy import interpolate
b = np.arange(0,13,1)
print(b)
print(a)
plt.plot(b,a)
f = interpolate.interp1d(b,a,'cubic')
c = np.arange(0,12,0.1)
print(c.shape)
d = f(c)
print(d)
d = d + np.random.randn(d.shape[0])*3
d = np.clip(d,0,100)
plt.plot(c,d)
```
